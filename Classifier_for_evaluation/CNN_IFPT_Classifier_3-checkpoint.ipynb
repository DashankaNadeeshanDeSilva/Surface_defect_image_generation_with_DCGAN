```python
from sklearn.model_selection import train_test_split
train_list, val_list = train_test_split(train_list, test_size=0.01)
```

```python
#data Augumentation
train_transforms =  transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
    ])

val_transforms = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
    ])


test_transforms = transforms.Compose([   
    transforms.Resize((224, 224)),
     transforms.RandomResizedCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor()
    ])
```

```python
class dataset(torch.utils.data.Dataset):
    
    def __init__(self,file_list,transform=None):
        self.file_list = file_list
        self.transform = transform
           
    #dataset length
    def __len__(self):
        self.filelength = len(self.file_list)
        return self.filelength
    
    #load an one of images
    def __getitem__(self,idx):
        img_path = self.file_list[idx]
        img = Image.open(img_path)
        img_transformed = self.transform(img)
        
        label = img_path.split('/')[-1].split('.')[0]
        # C:\\Project_arbeit\\GAN project\\Datasets\\IFPT_generated\\train\\train\\without_defects
        if label == 'C:\\Project_arbeit\\GAN project\\Datasets\\IFPT_generated\\train\\train\\with_defects':
            label=1
        elif label == 'C:\\Project_arbeit\\GAN project\\Datasets\\IFPT_generated\\train\\train\\without_defects':
            label=0
            
        return img_transformed,label
```

```python
train_data = dataset(train_list, transform=train_transforms)
test_data = dataset(test_list, transform=test_transforms)
val_data = dataset(val_list, transform=test_transforms)
```

```python
train_loader = torch.utils.data.DataLoader(dataset = train_data, batch_size=batch_size, shuffle=True )
test_loader = torch.utils.data.DataLoader(dataset = test_data, batch_size=batch_size, shuffle=True)
val_loader = torch.utils.data.DataLoader(dataset = val_data, batch_size=batch_size, shuffle=True)
```

```python
print(len(train_data), len(train_loader))
print(len(val_data), len(val_loader))
#check our images shape
train_data[0][0].shape
```

```python
# build Model
# 3 Convolution layer and 2 fully connected layer
# batchNormalization for limit overfitting

class Cnn(nn.Module):
    def __init__(self):
        super(Cnn,self).__init__()
        
        self.layer1 = nn.Sequential(
            nn.Conv2d(3,16,kernel_size=3, padding=0,stride=2),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        
        self.layer2 = nn.Sequential(
            nn.Conv2d(16,32, kernel_size=3, padding=0, stride=2),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2)
            )
        
        self.layer3 = nn.Sequential(
            nn.Conv2d(32,64, kernel_size=3, padding=0, stride=2),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )
        
        
        self.fc1 = nn.Linear(3*3*64,10)
        self.dropout = nn.Dropout(0.5)
        self.fc2 = nn.Linear(10,2)
        self.relu = nn.ReLU()
        
        
    def forward(self,x):
        out = self.layer1(x)
        out = self.layer2(out)
        out = self.layer3(out)
        out = out.view(out.size(0),-1)
        out = self.relu(self.fc1(out))
        out = self.fc2(out)
        return out
```

```python
model = Cnn().to(device)
model.train()
```

```python
optimizer = optim.Adam(params = model.parameters(),lr=0.001)
criterion = nn.CrossEntropyLoss()
```

```python
# Train the network

epochs = 25

for epoch in range(epochs):
    epoch_loss = 0
    epoch_accuracy = 0
    
    for data, label in train_loader:
        #print(label)
        #print(type(label))
        data = data.to(device)
        #label = torch.tensor(label).to(device)
        label = label.to(device)
        
        output = model(data)
        loss = criterion(output, label)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        acc = ((output.argmax(dim=1) == label).float().mean())
        epoch_accuracy += acc/len(train_loader)
        epoch_loss += loss/len(train_loader)
        
    print('Epoch : {}, train accuracy : {}, train loss : {}'.format(epoch+1, epoch_accuracy,epoch_loss))
    
    
    with torch.no_grad():
        epoch_val_accuracy=0
        epoch_val_loss =0
        for data, label in val_loader:
            data = data.to(device)
            label = label.to(device)
            
            val_output = model(data)
            val_loss = criterion(val_output,label)
            
            
            acc = ((val_output.argmax(dim=1) == label).float().mean())
            epoch_val_accuracy += acc/ len(val_loader)
            epoch_val_loss += val_loss/ len(val_loader)
            
        print('Epoch : {}, val_accuracy : {}, val_loss : {}'.format(epoch+1, epoch_val_accuracy,epoch_val_loss))
```

```python
with_defects_probs = []
model.eval()
with torch.no_grad():
    for data, fileid in test_loader:
        data = data.to(device)
        preds = model(data)
        preds_list = F.softmax(preds, dim=1)[:, 1].tolist()
        with_defects_probs += list(zip(list(fileid), preds_list))
        # print(f"predicted ----> {preds[0]}")
        
# For the best performance: if ground truith is with_defects, prediction probability should be 1

results_list = []
actual_classes_binary_list = []
predicted_classes_binary_list = []
count = 0

for i in range(len(with_defects_probs)):
    word_list = with_defects_probs[i][0].split('\\')
    
    # ground turth ----------
    ground_truth = word_list[-1]
    if ground_truth == 'with_defects':
        actual_class_binary = 1
    else: # case: actual_class = 'without_defects'
        actual_class_binary = 0
        
    # gather actual binary list
    actual_classes_binary_list.append(actual_class_binary)
    
    
    # predictions ----------
    pred_prob = with_defects_probs[i][1]
    if pred_prob >= 0.5:
        predicted_class = 'with_defects'
        predicted_class_binary = 1
    else:
        predicted_class = 'without_defects'
        predicted_class_binary = 0
        
    # gather predicted binary list
    predicted_classes_binary_list.append(predicted_class_binary)
        
    # Count how many correctly classfied
    if ground_truth == predicted_class:
        count = count + 1
        
    # result_temp = {'real': ground_truth, 'prediction': predicted_class, 'prediction prob': pred_prob}
    result_temp = {'real': ground_truth, 'prediction': predicted_class}
    results_list.append(result_temp) 

correctly_classified = count
accuracy =(correctly_classified / len(with_defects_probs))*100
print('Test accuracy on original images from IFPT dataset: ', accuracy,'%')    


# write results to a text file
text_file = open("C:\Project_arbeit\output.txt", "w")
text_file.write(str(accuracy))
text_file.close()
```

```python
# confusion matrix

import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt


# Data intake and confusion matrix calculation
data = {'y_Actual': actual_classes_binary_list,
        'y_Predicted': predicted_classes_binary_list
        }

df = pd.DataFrame(data, columns=['y_Actual', 'y_Predicted'])
confusion_matrix = pd.crosstab(df['y_Actual'], df['y_Predicted'], rownames=['Actual'], colnames=['Predicted'])
cf_matrix = confusion_matrix.to_numpy()

# Important stats
accuracy  = np.trace(cf_matrix) / float(np.sum(cf_matrix)) * 100
precision = cf_matrix[1,1] / sum(cf_matrix[:,1]) * 100
recall    = cf_matrix[1,1] / sum(cf_matrix[1,:]) * 100
f1_score  = 2*precision*recall / (precision + recall)
eval_stats = "\n\nAccuracy={:0.3f}\nPrecision={:0.3f}\nRecall={:0.3f}\nF1 Score={:0.3f}".format(accuracy,precision,recall,f1_score)

# Confustion matrix data process
group_names = ['True Neg', 'False Pos', 'False Neg', 'True Pos']
group_counts = ["{0:0.0f}".format(value) for value in
                cf_matrix.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in
                     cf_matrix.flatten() / np.sum(cf_matrix)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names, group_counts, group_percentages)]
labels = np.asarray(labels).reshape(2, 2)
categories = ['without defects', 'with defects']

# Confusion matrix create
sn.heatmap(confusion_matrix, annot=labels, fmt='', cmap='Blues', cbar=False, xticklabels=categories,yticklabels=categories)
plt.ylabel('Actual label')
plt.xlabel('Predicted label' + eval_stats)
plt.savefig(r'C:\Project_arbeit\project_to_REPORT\report images\confusion_matrix_2.jpg', dpi=300)
plt.show()


```
